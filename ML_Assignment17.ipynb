### 1. Using a graph to illustrate slope and intercept, define basic linear regression.

Basic linear regression is predicting the target variable by single feature. 

y= mx+c => m is the slope and c is the intercept.

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y= [6,5,4,3,2,1]
plt.plot(x,y)

slope=(y[1]-y[0])/(x[1]-x[0])
print("Slope =",slope)

intercept= y[0]-(slope*x[0])
print("Intercept = ", intercept)
```

### 2. What is the regression&#39;s standard error?
The regression standard error s a measure of the average distance between the actual values of the response variable (y) and the predicted values of the response variable (y). 


### 3. Provide an example of multiple linear regression.

Multiple linear regression involves finding the relationship between a dependent variable (Y) and two or more independent variables (X1, X2, ..., Xn).

Eg: Multiple linear regression model can be represented as:

Y = β0 + β1X1 + β2X2 + β3X3 + β4X4 + ε

Y is the selling price of the house, X1 is the area of the house, X2 is the number of bedrooms, X3 is the number of bathrooms, X4 is the age of the house, β0 is the intercept term, β1, β2, β3, and β4 are the regression coefficients for the independent variables, and ε is the error term.

### 5. Using an example, describe the polynomial regression model in detail.

Polynomial regression is a type of regression analysis in which the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth degree polynomial. This means that the relationship between X and Y is not linear.

Y = β0 + β1X + β2X^2 + β3X^3 + ... + βnX^n + ε

### 6. What are the logistic regression assumptions?

Logistic regression assumes that there is a linear relationship between the independent variables and the logit of the dependent variable and independent variables are independent of each other i.e. no multicollinearity.
